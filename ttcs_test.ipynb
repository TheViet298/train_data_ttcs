```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)`
from matplotlib import pyplot as plt
%matplotlib inline

# Read the CSV file with a specified encoding
file_path = '/kaggle/input/ttcs-total/Total.csv'
df = pd.read_csv(file_path, encoding='ISO-8859-1')

df.head()
```

```python
df.shape
```

```python
df['Country'].unique()
```

```python
# Kiểm tra các giá trị bị thiếu
print(df.isnull().sum())

# Xử lý giá trị thiếu
df.fillna(df.mean(), inplace=True)
```

```python
df.head
```

```python
df.head()
```

```python
# Loại bỏ các hàng bị trùng lặp
df = df.drop_duplicates()
```

```python
# Sử dụng One-Hot Encoding cho cột 'Continents'
df = pd.get_dummies(df, columns=['Continents'])
```

```python
print(df.columns)
```

```python
df.head()
```

```python
df = df.drop(columns=['STT', 'Economic Growth Rate of Education (% GDP)'])
```

```python
df.head()
```

```python
# Xử lý ngoại lai
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

# Loại bỏ các giá trị ngoại lai
df = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
# Loại bỏ các cột sau năm
df_cleaned = df.loc[:, :'Year']
```

```python
print(df_cleaned.head())
```

```python
df_cleaned.head()
```

```python
df.shape
```

```python
# Thay thế giá trị bằng 0 bằng giá trị trung bình của mỗi cột
df = df.replace(0, df.mean())
```
