```python
# Loại bỏ các hàng bị trùng lặp
df = df.drop_duplicates()
```

```python
# Sử dụng One-Hot Encoding cho cột 'Continents'
df = pd.get_dummies(df, columns=['Continents'])
```

```python
df = df.drop(columns=['STT'])
```

```python
df.head()
```

```python
df_cleaned = df.loc[:, :'Year']
df = df_cleaned
df.head()
```

```python
# Xử lý ngoại lai
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

# Loại bỏ các giá trị ngoại lai
df = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
df.shape
```

```python
# Kiểm tra các giá trị bị thiếu
print(df.isnull().sum())

# Xử lý giá trị thiếu
df.fillna(df.mean(), inplace=True)
```
